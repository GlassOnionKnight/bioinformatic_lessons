```python
def PatternCount(Text, Pattern):
    count = 0
    for i in range(len(Text) - len(Pattern) + 1):
        if Text[i:i+len(Pattern)] == Pattern:
            count += 1
    return count


text = "CGCGATACGTTACATACATGATAGACCGCGCGCGATCATATCGCGATTATC"
pattern = "CGCG"

print(PatternCount(text, pattern))

```
